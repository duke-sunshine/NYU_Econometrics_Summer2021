## Testing the Assumptions

### Assumption 1: Linear Model

In general, you use the scatter plot to inspect whether there is a linear relationship or not. 

### Assumption 2: The variance of the error term is constant. This variance does not depend on the values assumed by X.

```python
### get the residuals
predictions = model.predict(X)
residuals =y-predictions
df_res=pd.DataFrame({'Residual': residuals, 'Prediction': predictions})
df_res.head()
```

conclusions: it seems the variance of the error term is not constant. We might be able to improve the regression result by using an advanced General Linear Regression Model. 

```python
#print the scatter plot between residuals and predicitions
x='Prediction'
y='Residual'
sns.scatterplot(data=df_res, x=x, y=y)
```

It seems to represent a nonlinearity. In conclusion, we should consider a nonlinear model.

### Assumption 3: Normal Distribution: The error term is normally distributed For each fixed value of X, the distribution of Y is normal.

```python
# histogram of the residual
sns.displot(df_res, x="Residual")
```

```python
x=df_res['Residual']
pg.normality(x)
```

The formal test reject normality. We need to consider a different model. 

### Assumption 4: Collinearity: The independent variables in X are not correlated. This is no issues of multi-collinearity

```python
pg.pairwise_corr(df,columns=X.columns, method='pearson').round(3)
```

The result shows that transaction volumes are highly correlated with the following variables

FlowOutExUSD

FeeMedUSD

RevUSD

GasLmtTxMean.

In conclusion, we should drop those variables.

### Assumption 5: Zero Mean: The mean of the error term is 0.

```python
x=df_res['Residual']
pg.ttest(x, 0)
```

We cannot reject the null hypothesis that the error term has an expectation of zero. In conclusion, assumption 4 is likely to be satisfied.

##Logit Regression

```python
df['Positive_ROI30d'] = df['ROI30d'].apply(lambda x: 1 if x >0 else 0)
df[["Positive_ROI30d",'ROI30d']].head()
```

```python
y=df['Positive_ROI30d']
X=df[["TxTfrValAdjUSD","VtyDayRet180d","GasUsedTxMean"]]
```

```python
# add a constant to the dependent variables
X= sm.add_constant(X)
X.head()
```

```python
# conduct regression
logit_model = sm.Logit(y, X).fit()
```

```python
# print model summary
print(logit_model.summary())
```

```python
### get the odds ratio
np.exp(logit_model.params)
```

```python
### get the predictions 
pred_prob = logit_model.predict(X)
df_result=pd.DataFrame({'Pred_prob': pred_prob,'y':y})
df_result['Prediction']=df_result['Pred_prob'].apply(lambda x: 1 if x >=0.5 else 0)
df_result.head()
```

```python
from sklearn.metrics import accuracy_score
Prediction = df_result['Prediction']
accuracy_score(y, Prediction)
```

### Machine Learning

```python
## get the train and test data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2### get the predictions on the test data
pred_prob = logit_model.predict(X_test)
df_test=pd.DataFrame({'Pred_prob': pred_prob,'y_test':y_test})
df_test['Prediction']=df_test['Pred_prob'].apply(lambda x: 1 if x >=0.5 else 0)
df_test.head())
```

```python
# conduct regression
logit_model = sm.Logit(y_train, X_train).fit()
```

```python
# print model summary
print(logit_model.summary())
```

```python
### get the predictions on the test data
pred_prob = logit_model.predict(X_test)
df_test=pd.DataFrame({'Pred_prob': pred_prob,'y_test':y_test})
df_test['Prediction']=df_test['Pred_prob'].apply(lambda x: 1 if x >=0.5 else 0)
df_test.head()
```

```python
from sklearn.metrics import accuracy_score
Prediction =df_test['Prediction']
accuracy_score(y_test, Prediction)
```
