### Time Series Plots

```python
import plotly.express as px
import plotly.offline as py     
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
trace0 = go.Scatter(x = df["time"], y = df["AdrBalUSD100Cnt"], hoverinfo='x+y', mode='lines',  name = 'Addresses, with balance, greater than $100, count')
trace1 = go.Scatter(x = df["time"], y = df["AdrBalUSD100KCnt"], hoverinfo='x+y', mode='lines',  name = 'Addresses, with balance, greater than $100K, count')
trace2 = go.Scatter(x = df["time"], y = df["AdrActCnt"], hoverinfo='x+y', mode='lines',  name = 'Total Active Address, count')
layout = go.Layout(title="Distribution of Address with Different Balance Floor ", xaxis = dict(title="date"), yaxis = dict(title ="count") ) 
data = [trace0, trace1,trace2]
fig1 = go.Figure(layout = layout, data = data) 
fig1.update_yaxes(type="log")
fig1.update_xaxes(
    rangeslider_visible=True,
    rangeselector=dict(
        buttons=list([
            dict(step="all"),
            dict(count=1, label="1m", step="month", stepmode="backward"),
            dict(count=3, label="3m", step="month", stepmode="backward"),
            dict(count=3, label="6m", step="month", stepmode="backward"),
            dict(count=1, label="1y", step="year", stepmode="backward"),
            dict(count=1, label="YTD", step="year", stepmode="todate"),
          
        ])))
py.iplot(fig1)
```

# Regression

https://www.statsmodels.org/stable/regression.html

```python
import statsmodels.api as sm
```

```python
# define the dependent and independent variables
X=df[['AdrActCnt','NVTAdj90']]
y=df['PriceUSD']
```

```python
# add a constant to the dependent variables
X= sm.add_constant(X)
X.head()
```

```python
# conduct regression
model = sm.OLS(y, X).fit()
```

```python
# print model summary
print(model.summary())
```
